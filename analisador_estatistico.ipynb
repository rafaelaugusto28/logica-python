```python
alunos = [
    {'nome': 'João Silva', 'notas': [8.5, 7.0, 9.0]},
    {'nome': 'Ana Souza', 'notas': [5.5, 10.0, 9.5]},
    {'nome': 'Carlos Oliveira', 'notas': [4.0, 3.5, 5.0]},
    {'nome': 'Mariana Costa', 'notas': [10.0, 10.0, 10.0]},
    {'nome': 'Ricardo Santos', 'notas': [7.0, 6.5, 7.5]},
    {'nome': 'Beatriz Lima', 'notas': [2.0, 4.0, 3.0]},
    {'nome': 'Fernando Gomes', 'notas': [9.0, 8.5, 8.0]},
    {'nome': 'Julia Martins', 'notas': [6.0, 6.0, 6.5]},
    {'nome': 'Lucas Pereira', 'notas': []}, 
    {'nome': 'Patrícia Rocha', 'notas': [8.0, 9.0, 7.5]}
]
```

```python
def calcular_media_aluno(notas):
    try:
        media = sum(notas) / len(notas)
        return media
    except ZeroDivisionError:
        return 0.0
   
```

```python
def classificar_aluno(media):
   return "Aprovado" if media >= 7 else "Reprovado"
```

```python
def calcular_media_turma(alunos):
    soma_total_das_medias = 0
    
    for aluno in alunos:
        media_individual = calcular_media_aluno(aluno['notas'])
        soma_total_das_medias += media_individual
    
    media_geral = soma_total_das_medias / len(alunos)
    return media_geral
```

```python
for aluno in alunos:
    m = calcular_media_aluno(aluno['notas'])
    c = classificar_aluno(m)
    print(f"O aluno {aluno['nome']} ficou com média {m:.1f} e está {c}")

media_sala = calcular_media_turma(alunos)
print(f"\n--- Média Geral da Turma: {media_sala:.1f} ---")
```
