```python
diametro_laranja = dados[:5000, 0] # Linhas :5000, Coluna 0
peso_laranja     = dados[:5000, 1] # Linhas :5000, Coluna 1

# Toranja (Linhas 5000 at√© o fim)
diametro_toranja = dados[5000:, 0] # Linhas 5000:, Coluna 0
peso_toranja     = dados[5000:, 1] # Linhas 5000:, Coluna 1
```

```python
import matplotlib.pyplot as plt  #para fazer graficos
```

LARANJA:



```python
n=np.size(diametro_laranja)
x=diametro_laranja
y=peso_laranja
```

```python
a = (n*np.sum(x*y)-np.sum(x)*np.sum(y))/(n*np.sum(x**2)-np.sum(x)**2)
```

```python
b=np.mean(y)-a*np.mean(x)
```

```python
y=a*x+b
```

```python
plt.plot(diametro_laranja,peso_laranja,label='Laranja',color='orange')
plt.plot(x,y)
```

TORANJA:

```python
N=np.size(diametro_toranja)
X=diametro_toranja
Y=peso_toranja
```

```python
A = (N*np.sum(X*Y)-np.sum(X)*np.sum(Y))/(N*np.sum(X**2)-np.sum(X)**2)
```

```python
B=np.mean(Y)-a*np.mean(X)
```

```python
Y=A*X+B
```

```python
plt.plot(diametro_toranja,peso_toranja,label='Toranja',color='red')
plt.plot(X,Y)
```
